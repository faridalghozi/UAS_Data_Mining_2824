### 1. Linear Regression
* <pre>Independent Variable (X) = ("model", "year", "transmission", "mileage", "fuelType", "tax", "mpg", "engineSize")</pre>
* <pre>Dependent Variable (y) = ("price")</pre>

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
df_car = pd.read_csv('toyota.csv')
df_car
```

```python
df_car.info()
```

```python
car2 = pd.get_dummies(df_car)
car2
```

```python
X = car2.drop(columns="price")
y = car2["price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=10)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print(linreg.coef_)
print(linreg.intercept_)
```

```python
pred = linreg.predict(X_test)
print(pred)
print(y_test)
```

```python
evaluasi = pd.DataFrame({'Actual' : y_test, 'Predicted' : pred})
evaluasi
```

```python
linreg.score(X_test, y_test)
```

```python
print('Mean Absolute Error :', mean_absolute_error(y_test, pred))
print('Mean Squared Error :', mean_squared_error(y_test, pred))
print('Root Mean Squared Error :', mean_squared_error(y_test, pred, squared=False))
```

### 2. Linear Regression
* <pre>Independent Variable (X) = ("year", "mileage", "tax", "mpg", "engineSize")</pre>
* <pre>Dependent Variable (y) = ("price")</pre>

```python
df_mobil = pd.read_csv('toyota.csv')
df_mobil
```

```python
df_mobil.info()
```

```python
X = df_mobil.drop(columns=["price", "model", "transmission", "fuelType"])
X = X.values
y = df_mobil["price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=10)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print(linreg.coef_)
print(linreg.intercept_)
```

```python
pred2 = linreg.predict(X_test)
print(pred2)
print(y_test)
```

```python
evaluasi2 = pd.DataFrame({'Actual' : y_test, 'Predicted' : pred2})
evaluasi2
```

```python
linreg.score(X_test, y_test)
```

```python
print('Mean Absolute Error :', mean_absolute_error(y_test, pred2))
print('Mean Squared Error :', mean_squared_error(y_test, pred2))
print('Root Mean Squared Error :', mean_squared_error(y_test, pred2, squared=False))
```
